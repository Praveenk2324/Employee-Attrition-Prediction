```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv(r"..\data\raw\HR-Employee-Attrition.csv")
df.head()
```

```python
df.info()
```

```python
# Calculates the percentage of each unique value in the 'Attrition' column
percentages = df['Attrition'].value_counts(normalize=True) * 100

print(percentages)
```

Plot distributions of Age, MonthlyIncome, JobSatisfaction vs attrition

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Age', hue='Attrition', kde=True, bins=30, palette='Set2')
plt.title('Distribution of Age by Attrition')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', palette='Set2')
plt.title('Monthly Income vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='MonthlyIncome', hue='Attrition', kde=True, bins=30, palette='Set2') 
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='JobSatisfaction', hue='Attrition', palette='Set2')
plt.title('Job Satisfaction vs Attrition')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Count of Employees')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Age (Histogram)
sns.histplot(data=df, x='Age', hue='Attrition', kde=True, ax=axes[0], palette='Set1')
axes[0].set_title('Age vs Attrition')

# 2. Monthly Income (Boxplot)
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', ax=axes[1], palette='Set1')
axes[1].set_title('Monthly Income vs Attrition')

# 3. Job Satisfaction (Countplot)
sns.countplot(data=df, x='JobSatisfaction', hue='Attrition', ax=axes[2], palette='Set1')
axes[2].set_title('Job Satisfaction vs Attrition')

plt.tight_layout()
plt.show()
```
